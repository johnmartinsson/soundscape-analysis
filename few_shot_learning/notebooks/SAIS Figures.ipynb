```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.boxplot([multi_val])
plt.axhline(40, c='black')
plt.legend(['Baseline'])
plt.ylabel('F-measure')
plt.savefig('SAIS fig/multi_val.pdf')
```
