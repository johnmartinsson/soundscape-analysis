```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import math
```

Works with output from experiments/negfraction_clusterK_map.py

```python
def heatmap(path_scores, num_elements):
    
    file = open(path_scores, 'rb')
    scores = pickle.load(file)
    
    data = np.zeros(num_elements)
    for i, key in enumerate(scores.keys()):
        tmp_f = []
        for inner in scores[key].keys():
            tmp_f += [scores[key][inner]['fmeasure (percentage)']]
        data[i] = np.array(tmp_f).mean()
        
    a = np.empty(len(list(scores.keys())), dtype=object)
    a[:] = list(scores.keys())
    for i in range(len(a)):
        a[i] = str(a[i])
    a = a.reshape((int(math.sqrt(num_elements)),int(math.sqrt(num_elements))))
    
    ax = sns.heatmap(np.reshape(data, (5,5)), linewidth=0.5, annot=a, fmt='')
    frac_range = [0.2,0.4,0.6,0.8,1]
    k_range = [1,3,5,7,9]
    #ax.xaxis(frac_range)
```

```python
val_scores = '/home/willbo/repos/soundscape-analysis/few_shot_learning/experiments/map_specaug_resnet_semi5all_spl_seed100_VAL/scores.pkl'
test_scores = '/home/willbo/repos/soundscape-analysis/few_shot_learning/experiments/map_specaug_resnet_semi5all_spl_seed100_TEST/scores.pkl'
```

```python
heatmap(val_scores, 25)
```
