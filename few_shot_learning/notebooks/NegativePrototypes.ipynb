```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
pklfile = open('../experiments/randomepisode_valset_allpos_varyneg/scores.pkl', 'rb')
scores = pickle.load(pklfile)
pklfile.close()
```

```python
m_f = np.zeros((len(scores.keys()),
             len(scores[list(scores.keys())[0]].keys())))
m_p = np.zeros(m_f.shape)
m_r = np.zeros(m_f.shape)

for i, key in enumerate(scores.keys()):
    tmp_f = []
    tmp_p = []
    tmp_r = []
    for skey in scores[key].keys():
        tmp_f += [scores[key][skey]['fmeasure (percentage)']]
        tmp_p += [scores[key][skey]['precision']]
        tmp_r += [scores[key][skey]['recall']]
    m_f[i] = np.array(tmp_f)
    m_p[i] = np.array(tmp_p)
    m_r[i] = np.array(tmp_r)
```

```python
plt.figure(figsize=(12,8))
fig = plt.boxplot(m_f.T, labels=list(scores.keys()))
plt.xlabel('Fraction')
plt.ylabel('F measure percentage')
#plt.savefig('fmeasure_varypos.pdf')
```

```python
plt.figure(figsize=(12,8))
fig = plt.boxplot(m_p.T, labels=list(scores.keys()))
plt.xlabel('Fraction')
plt.ylabel('Precision')
```

```python
plt.figure(figsize=(12,8))
fig = plt.boxplot(m_r.T, labels=list(scores.keys()))
plt.xlabel('Fraction')
plt.ylabel('Recall')
```
